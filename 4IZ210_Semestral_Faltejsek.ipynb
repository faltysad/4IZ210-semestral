```python
def tf(corpus, query = None):
    tfs = []

    for document in corpus:
        words = [w.lower() for w in document.split()]
        dict={}
        
        if query:
          queryTerms = [q.lower() for q in query.split()]
          words = [a for a in words if a in queryTerms]

        for word in words:
            dict[word] = dict[word]+1 if word in dict else 1
        tfs.append(dict)
    return tfs
```

\begin{align}
    \text{Normalizovaná frekvence }tf_{i,j} \text{ termu } t_i \text{ v dokumentu } D_j:\\
    tf_{ij} = \frac{f_{i,j}}{max_j}
    \text{ kde: } max_j=max{f_{1,j},f_{2,j}, ..., f_{K,j}} \\
    0\le tf_{i,j} \le 1
\end{align}

```python
def normalized_tf(corpus, query = None):
    tfs = []

    for document in corpus:
        words = [w.lower() for w in document.split()]
        dict={}
        
        if query:
          queryTerms = [q.lower() for q in query.split()]
          words = [a for a in words if a in queryTerms]

        for word in words:
            dict[word] = dict[word]+1 if word in dict else 1

        max_frequency = max(dict.values())
        for word,freq in dict.items():
            dict[word]=freq/max_frequency
        tfs.append(dict)
    return tfs
```

\begin{align}
  \text{Inverzní dokumentová frekvence } idf_i \text{ termu } t_i \text{ v } {D_i, ..., D_N} \\
  df_i = \text{počet dokumentů, ve kterých se term } t_i \text{ vyskytuje alespoň jednou} \\
  idf_i = log(\frac{N}{df_i})
\end{align}

```python
def idf(df, N):
  return math.log10((N / df))

def sum_row(row):
  occurences = 0
  
  print(row)
  
  return occurences

def get_dataframe_with_df_idf_columns(term_frequency, N):  
  dataframe = pd.DataFrame.from_dict(term_frequency).fillna(0);

  dataframe = dataframe.T

  dataframe["df"] = dataframe[[0, 1, 2, 3, 4]].apply(lambda x: (x>0).sum(), axis=1)
  dataframe["idf"] = dataframe["df"].apply(lambda row: idf(row, N))

  return dataframe
```

\begin{align}
  \text{TF-IDF = Term Frequency - Inverse Document Frequency} \\
  \text{TF-IDF váha }w_{i,j} \text{ termu } t_i \text{ pro dokument } D_i: \\
  w_{i,j} = tf_{i,j}*idf_i
\end{align}

## 4a) Hodnoty TF-IDF pro každý z dokumentů

```python
df_idf_terms_weight_dataframe = normalized_term_frequency_dataframe.multiply(df_idf_dataframe["idf"], axis="index")
df_idf_terms_weight_dataframe
```

```python
def get_query_vector_dataframe(term_frequency_dataframe, idf, query):
  queryTerms = [q.lower() for q in query.split()]
  terms = term_frequency_dataframe.index

  terms_weights = {}

  for term in terms:
    if term in queryTerms:
      terms_weights[term] = 1
    else:
      terms_weights[term] = 0
  

  
  vector_dataframe = pd.DataFrame.from_dict([terms_weights]).T
  vector_dataframe["idf"] = idf
  vector_dataframe["q"] = vector_dataframe[0].multiply(vector_dataframe["idf"], axis="index")

  return vector_dataframe
```

## 4b) Hodnoty TF-IDF pro dotaz

```python
query_vector_dataframe = get_query_vector_dataframe(term_frequency_dataframe, df_idf_dataframe["idf"], "herní i5 ssd nvidia DDR4")
query_vector_dataframe
```

\begin{align}
  \text{Míra podbnosti dotazu a dokumentu} \\
  Dokument D_j = <w_{1,j}, ..., w_{1,K}> \\
  Dotaz Q = <q_1, ..., q_K> \\
  \\
  \text{Kosinová míra podobnosti:} \frac{\sum_{i=i}^K w_{i,j} \times q_i}{{\sqrt{\sum_{i=i}^K w_{i,j}^2}} \times \sqrt{\sum_{i=i}^K q_i^2}}
\end{align}

**Kosinová míra podobnosti se použije pro:**
* sestupné uspořádání vyhledaných dokumetnů


```python
def cosine_similarity(a, b):
  cos_sim = dot(a, b)/(norm(a)*norm(b))

  return cos_sim
```

## 5) Hodnoty kosinové míry podobnosti dotazu s každým dokumentem.

```python
q = query_vector_dataframe["q"]
D1_tf_idf_weight = df_idf_terms_weight_dataframe[0]
D2_tf_idf_weight = df_idf_terms_weight_dataframe[1]
D3_tf_idf_weight = df_idf_terms_weight_dataframe[2]
D4_tf_idf_weight = df_idf_terms_weight_dataframe[3]
D5_tf_idf_weight = df_idf_terms_weight_dataframe[4]

D1_cosine_similarity_dict = { "name": "D1", "similarity": cosine_similarity(D1_tf_idf_weight, q) }
D2_cosine_similarity_dict = { "name": "D2", "similarity": cosine_similarity(D2_tf_idf_weight, q) }
D3_cosine_similarity_dict = { "name": "D3", "similarity": cosine_similarity(D3_tf_idf_weight, q) }
D4_cosine_similarity_dict = { "name": "D4", "similarity": cosine_similarity(D4_tf_idf_weight, q) }
D5_cosine_similarity_dict = { "name": "D5", "similarity": cosine_similarity(D5_tf_idf_weight, q) }

results = [D1_cosine_similarity_dict, D2_cosine_similarity_dict, D3_cosine_similarity_dict, D4_cosine_similarity_dict, D5_cosine_similarity_dict]
sorted_results = sorted(results, key=lambda d: d['similarity'], reverse=True)

pd.DataFrame.from_dict(sorted_results)
```
